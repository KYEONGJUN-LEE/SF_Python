## PyTorch Quiz 1 : Tensor 초기화

````python
import torch
import numpy as np
````

### 1. 빈 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 3x4 크기의 Tensor
- 값이 초기화되지 않은 상태

````python
tensor = torch.empty(3,4)
print(tensor)
````

### 2. 0으로 채워진 Tensor 생성하기
다음과 같은 Tensor를 생성하세요.
- 2x5x3 크기의 Tensor
- 데이터 타입은 float64

````python
a = torch.zeros([2,5,3],dtype=torch.float64)
print(a)
````

### 3. 1로 채워진 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 4x4 크기의 Tensor
- 데이터 타입은 int

````python
a = torch.ones(4,4,dtype=torch.int)
print(a)
````

### 4. 랜덤 값으로 채워진 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 3x3 크기의 Tensor
- 모든 값이 0과 1 사이의 랜덤 값

````python
a = torch.rand(3,3,dtype=torch.float)
print(a)
````

### 5. 다차원 Tensor 생성하기
다음과 같은 다차원 Tensor를 생성하세요.
- 값이 다음과 같은 2x3의 2D 배열
    ```
    [[1, 2, 3],
     [4, 5, 6]]
    ```

````python
nd_arr = np.array([[1,2,3],[4,5,6]])
print(nd_arr)

a = torch.from_numpy(nd_arr)
print(a)
````

### 6. Tensor의 값을 변경하기
1. 표준정규분포의 값을 랜덤으로 갖는 2x2 Tensor를 생성하세요.
2. 생성된 Tensor의 값을 모두 0으로 바꿔보세요.

````python
tensor = torch.randn(2, 2)
result = torch.zeros_like(tensor)
print(result)
````
