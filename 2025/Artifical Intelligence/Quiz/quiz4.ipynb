### 문제9
- 텐서 [[1, 2], [3, 4], [5, 6]]에서:
- dim=1 기준으로 크기 1씩 나누세요.
- 각각의 조각을 새로운 축에 쌓은 후 출력해보세요.

````python
# 여기에 코드를 작성합니다
x = torch.tensor([[1, 2], [3, 4], [5, 6]])
# 1. 열 기준으로 크기 1씩 나누기
y = torch.split(x,1,dim=1)
print(y)
# 2. 새로운 축에 쌓기
result = torch.stack(y)
print(result)
````

### 문제10
- 다음 텐서를 활용해서 아래 문제를 해결해보세요
  ```
  a = torch.tensor([[1, 2], [3, 4]])
  b = torch.tensor([[5, 6], [7, 8]])
  ```

1. 두 텐서를 열 기준으로 연결하고 결과를 출력하세요.
2. 연결된 텐서를 크기 `(4, 2)`로 변환하세요.
3. 변환된 텐서를 2개의 조각으로 나누고, 각 조각을 0번 축으로 스택하여 출력하세요.


````python
# 여기에 코드를 작성합니다
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6], [7, 8]])
# 1. 열 기준으로 병합
cat = torch.cat((a,b),dim = 1)
print(cat)
# 2. 크기 (4, 2)로 변환
view = cat.view(4,2)
print(view)
# 3. 2개의 조각으로 나누고
chunk = torch.chunk(view,2)
print(chunk)
# 4. 0번 축으로 스택
result = torch.stack(chunk,dim=0)
print(result)
````
